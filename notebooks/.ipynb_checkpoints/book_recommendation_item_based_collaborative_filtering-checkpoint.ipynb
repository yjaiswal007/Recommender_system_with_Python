### Collaborative filtering using K-Nearest Neighbors

#### Some data processing

```python
# Consider those users who have rated at least 200 books and those ratings which got at least 100 counts
counts1 = df_ratings['userID'].value_counts()
df_ratings = df_ratings[df_ratings['userID'].isin(counts1[counts1 >= 200].index)]
counts = df_ratings['bookRating'].value_counts()
df_ratings = df_ratings[df_ratings['bookRating'].isin(counts[counts >= 100].index)]
df_ratings.shape
```

```python
df_combine_book_rating = pd.merge(df_ratings, df_books[['ISBN', 'bookTitle']], on='ISBN', how='left')
display(df_combine_book_rating.head())
df_combine_book_rating.shape
```

```python
df_combine_book_rating.isnull().sum()
```

```python
df_combine_book_rating[df_combine_book_rating.bookTitle.isna()].head()
```

```python
%%time
df_ratingCounts = df_combine_book_rating.groupby('ISBN').apply(lambda x: pd.Series({
    'mean_rating': x['bookRating'].mean(),
    'totalRatingCount': x['bookRating'].count()
}), include_groups=False).reset_index()

df_ratingCounts.head()
```

```python
df_combine_book_rating = df_combine_book_rating.merge(df_ratingCounts, on='ISBN', how='left')
df_books_with_rating = df_combine_book_rating[df_combine_book_rating.totalRatingCount >= 50]
```

```python
df_books_with_rating.head()
```

```python
df_books_with_rating.shape
```

#### KNN

```python
df_books_with_rating = df_books_with_rating.drop_duplicates(['userID', 'ISBN'])
df_books_with_rating_pivot = df_books_with_rating.pivot(index = 'ISBN', columns = 'userID', values = 'bookRating').fillna(0)
df_books_with_rating_matrix = csr_matrix(df_books_with_rating_pivot.values)
```

```python
df_books_with_rating_pivot.head(2)
```

```python
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(df_books_with_rating_matrix)
```

#### Choose a random title and recommends related books

```python
random_index = np.random.choice(df_books_with_rating.shape[0])
chosen_ISBN = df_books_with_rating.iloc[random_index].ISBN
chosen_Title = df_books_with_rating.iloc[random_index].bookTitle    

print("Searching books based on \"%s\""%chosen_Title)
```

```python
distances, indices = model_knn.kneighbors(df_books_with_rating_pivot.loc[chosen_ISBN].values.reshape(1,-1), n_neighbors = 6)
```

```python
rec_ISBN = df_books_with_rating_pivot.iloc[indices.flatten()].index.values
df_recommendation = pd.DataFrame(zip(rec_ISBN, distances.flatten()), columns=['ISBN', 'Cosine_distance'])
df_recommendation = df_recommendation.merge(df_books_with_rating[['ISBN', 'bookTitle', 'mean_rating', 'totalRatingCount']], on='ISBN', how='left')
df_recommendation.drop_duplicates(subset='ISBN', inplace=True)
df_recommendation = df_recommendation[df_recommendation.bookTitle!=chosen_Title]
```

```python
print('Here\'s the list of books similar to '+'\033[1m'+str(chosen_Title)+'\033[0m'+'.\n')
display(df_recommendation[['Cosine_distance', 'ISBN', 'bookTitle', 'mean_rating', 'totalRatingCount']])
```
